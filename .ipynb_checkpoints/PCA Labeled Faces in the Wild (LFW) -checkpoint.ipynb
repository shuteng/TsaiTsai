```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_lfw_people
```

```python
data = fetch_lfw_people()
```

```python
x = data.data
```

```python
x.shape
```

```python
data.images.shape
```

```python
plt.figure(figsize=[10,5])

for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(data.images[i], cmap='gray')
    plt.axis('off')    
plt.show()
```

```python
pca = PCA(n_components=100).fit(x)
V= pca.components_
pca.explained_variance_ratio_
```

```python
x_new = pca.transform(x)
```

```python
x_new.shape
```

```python
plt.figure(figsize=[10,5])
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(V[i].reshape(62,47), cmap='gray')
    plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=[10,5])
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(V[i+50].reshape(62,47), cmap='gray')
    plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=[10,5])
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(x_new[i].reshape(10,10), cmap='gray')
    plt.axis('off')
plt.show()
```

```python
x_inverse = pca.inverse_transform(x_new)
```

```python
x_inverse.shape
```

```python
plt.figure(figsize=[15,5])
for i in range(10):
    plt.subplot(2,10,i+1)
    plt.imshow(data.images[i], cmap='gray')
    plt.axis('off')
    plt.subplot(2,10,i+11)
    plt.imshow(x_inverse[i].reshape(62,47), cmap='gray')
    plt.axis('off')
plt.show()
```
