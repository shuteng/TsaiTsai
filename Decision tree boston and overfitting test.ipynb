```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.datasets import load_boston
```

```python
data = load_boston()
```

```python
data.data.shape
```

```python
data.feature_names
```

```python
reg = DecisionTreeRegressor(random_state=1)
score = cross_val_score(reg, data.data, data.target, cv = 5)
```

```python
rng = np.random.RandomState(1)
```

```python
x = np.sort(5* rng.rand(80,1))
y = np.sin(x).ravel()
y[::5] += 3* (0.5- rng.rand(16))
```

```python
reg1 = DecisionTreeRegressor(max_depth=3)
reg2 = DecisionTreeRegressor(max_depth=5)
reg1.fit(x,y)
reg2.fit(x,y)
```

```python
x_test = np.arange(0,5,0.01).reshape(-1,1)
y1 = reg1.predict(x_test)
y2 = reg2.predict(x_test)
```

```python
plt.figure(figsize=(15,8))
plt.scatter(x,y, s=20, c = 'darkorange', label = 'data')
plt.plot(x_test, y1, color = 'yellowgreen', label = 'max_depth=3')
plt.plot(x_test, y2, color = 'cornflowerblue', label = 'max_depth=5')
plt.xlabel('data')
plt.ylabel('target')
plt.title('Decision tree regression')
plt.legend()
plt.show()
```
