```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
x = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.DataFrame(data.target, columns=['target'])
```

```python
x.shape
```

```python
y.shape
```

```python
pca = PCA(n_components=2)
pca = pca.fit(x)
x_de = pca.transform(x)
x_y = pd.concat([pd.DataFrame(x_de),y], axis=1)
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
pca.components_.shape
```

```python
plt.figure(figsize=[7,5])
color=['red', 'blue', 'green']
for i in x_y['target'].unique():
    plt.scatter(x_y.loc[x_y['target']==i, 0], x_y.loc[x_y['target']==i, 1], alpha=0.3, c=color[i], label = data.target_names[i])
plt.legend()
plt.title('PCA of Iris')
plt.xlabel('PC1 - %.4f'%pca.explained_variance_ratio_[0])
plt.ylabel('PC2 - %.4f'%pca.explained_variance_ratio_[1])
plt.show()
```
